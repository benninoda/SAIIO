```python
import numpy as np
```

```python
class ResourceAllocationProblemSolver:
    def __init__(self, A):
        self.A = A
        self.K = A.shape[0]
        self.Q = A.shape[1]
        self.profit_matrix = np.append(A[0].copy().reshape(1, -1), np.zeros((self.K - 1, self.Q)), axis=0)
        self.allocation_matrix = np.append(np.array(list(range(self.Q))).reshape(1, -1), np.zeros((self.K - 1, self.Q)), axis=0)
        
    def calculate_max_profit(self, k, q):
        current_profit = self.A[k][0] + self.profit_matrix[k - 1][0]
        current_resource = 0
        
        for i in range(q + 1):
            other_resources = q - i
            print(f'for current agent {i} other_resources {other_resources}')
            other_profit = self.profit_matrix[k - 1][other_resources]
            print(f"now + other {self.A[k][i]} + {other_profit} current {current_profit}")
            if self.A[k][i] + other_profit > current_profit:
                current_profit = self.A[k][i] + other_profit
                current_resource = i
        return current_profit, current_resource
        
    def solve(self):
        for k in range(1, self.K):
            for q in range(self.Q):
                print(f"agents {k + 1} resources {q}")
                current_profit, current_resource = self.calculate_max_profit(k, q)
                self.profit_matrix[k][q] = current_profit
                self.allocation_matrix[k][q] = current_resource
                print(self.profit_matrix)
                print(self.allocation_matrix)
        q = self.Q - 1
        resolved = np.zeros(self.K)
        for i in range(self.K):
            resolved[self.K - i - 1]  = self.allocation_matrix[self.K - i - 1][int(q)]
            q -=  resolved[self.K - i - 1]
            
        return self.profit_matrix[self.K - 1, self.Q - 1], resolved
        
```

```python
A = np.array([
    [0, 1, 2, 3],
    [0, 0, 1, 2],
    [0, 2, 2, 3]
])
```

```python
solver = ResourceAllocationProblemSolver(A)
print(solver.solve())
```
